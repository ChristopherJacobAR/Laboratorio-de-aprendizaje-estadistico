```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression, Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
df=pd.read_excel("Motor Trend Car Road Tests (1).xlsx").dropna()
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df_numeric = df.drop(columns=["model"])

X = df_numeric.drop(columns=["mpg"])   
y = df_numeric["mpg"]                 
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.60, random_state=1234
)

model = LinearRegression()
model.fit(X_train, y_train)

r2_train = model.score(X_train, y_train)
r2_test = model.score(X_test, y_test)

r2_test,r2_train
```

```python
pd.DataFrame({
    "Variable": X_train.columns,
    "Beta": model.coef_
}).sort_values(by="Beta", ascending=False)
```

```python
from sklearn.linear_model import Ridge
```

```python
alphas = [0.01, 0.1, 1, 10, 100]

results = []

for alpha in alphas:
    ridge = Ridge(alpha=alpha)
    ridge.fit(X_train, y_train)
    
    r2_train = ridge.score(X_train, y_train)
    r2_test  = ridge.score(X_test, y_test)
    
    results.append({
        "alpha": alpha,
        "R2_Train": r2_train,
        "R2_Test": r2_test
    })

df_results = pd.DataFrame(results)
df_results
```

```python

X_qsec = df_numeric.drop(columns=["qsec"])
y_qsec = df_numeric["qsec"]

X_train_q, X_test_q, y_train_q, y_test_q = train_test_split(
    X_qsec, y_qsec, test_size=0.60, random_state=42
)

lr_q = LinearRegression()
lr_q.fit(X_train_q, y_train_q)

r2_train_q = lr_q.score(X_train_q, y_train_q)
r2_test_q  = lr_q.score(X_test_q, y_test_q)

print("Regresión lineal sin regularización:")
print("R2 Entrenamiento:", r2_train_q)
print("R2 Prueba:", r2_test_q)

alphas = [0.01, 0.1, 1, 10, 100]
results_qsec = []

for alpha in alphas:
    ridge_q = Ridge(alpha=alpha)
    ridge_q.fit(X_train_q, y_train_q)
    
    r2_train_ridge = ridge_q.score(X_train_q, y_train_q)
    r2_test_ridge  = ridge_q.score(X_test_q, y_test_q)
    
    results_qsec.append({
        "alpha": alpha,
        "R2_Train": r2_train_ridge,
        "R2_Test": r2_test_ridge
    })

df_results_qsec = pd.DataFrame(results_qsec)
df_results_qsec
```

```python
df_dummies = pd.get_dummies(df.drop(columns=["model"]),
                            columns=["cyl", "gear", "carb"],
                            drop_first=True)

X = df_dummies.drop(columns=["mpg"])
y = df_dummies["mpg"]

lr = LinearRegression()
lr.fit(X, y)

r2_global = lr.score(X, y)

coef_df = pd.DataFrame({
    "Variable": X.columns,
    "Beta": lr.coef_
}).sort_values(by="Beta", ascending=False)

print("Regresión lineal (todos los datos):")
print("R² global:", r2_global)
print("\nCoeficientes del modelo:")
print(coef_df)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.60, random_state=42)

lr_split = LinearRegression()
lr_split.fit(X_train, y_train)

r2_train = lr_split.score(X_train, y_train)
r2_test  = lr_split.score(X_test, y_test)

print("\nRegresión lineal (con train-test-split):")
print("R² de entrenamiento:", r2_train)
print("R² de prueba:", r2_test)
alphas = [0.01, 0.1, 1, 10, 100]
results = []

for alpha in alphas:
    ridge = Ridge(alpha=alpha)
    ridge.fit(X_train, y_train)
    
    results.append({
        "alpha": alpha,
        "R2_Train": ridge.score(X_train, y_train),
        "R2_Test": ridge.score(X_test, y_test)
    })

df_results = pd.DataFrame(results)
print("\nResultados con regularización L2 (Ridge):")
df_results
```
